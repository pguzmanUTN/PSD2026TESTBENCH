```python
#Imports
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sp
```

```python
#Funcion del generador de se;ales
def  mi_funcion ( vmax = 1, dc = 0, ff = 1, ph = 0, nn = 100, fs = 50, type_s = 'sine', duty = 0.5) :
    
    tt = np.arange(start = 0, step = 1/fs, stop = N/fs)
    if   type_s == 'sine' :
        xx = vmax*np.sin(2*np.pi*ff*tt+ph)+dc
    elif type_s == 'sawtooth' :
        xx = vmax*sp.sawtooth(2*np.pi*ff*tt+ph)+dc
    elif type_s == 'triangle' :
        xx = vmax*sp.sawtooth(2*np.pi*ff*tt+ph,width = 0.5)+dc
    elif type_s == 'invsawtooth' :
        xx = vmax*sp.sawtooth(2*np.pi*ff*tt+ph, width = 0)+dc
    elif type_s == 'square' :
        xx = vmax*sp.square(2*np.pi*ff*tt+ph, duty = duty)+dc
    
    return tt, xx
```

```python
#Parametros del muestreo
N    = 1000 #Muestras
fs   = N #Resolucion espectral normalizada
#Parametros de la Senal
ff   = 10 # Hz
vmax = 5 #Volt
dc   = 1 #Volt
px   = np.pi/2 #rads
type_s = 'sine'
duty = 0.75 #Solo para senal cuadrada
```

```python
tt, xx = mi_funcion( vmax = vmax, dc = dc, ff = ff, ph= px, nn = N, fs = fs, type_s = type_s)
plt.xlim([0,5/ff]) #Para que me muestre 5 ciclos
plt.plot(tt, xx, marker = 'X', color = 'r', linestyle = '-')
plt.grid()
plt.title("Salida Generador de Senales")
plt.ylabel("Amplitud [V]")
plt.xlabel("Tiempo [s]")
```

```python
tt, xx = mi_funcion( vmax = vmax, dc = dc, ff = ff, ph= px, nn = N, fs = fs, type_s = 'square', duty = duty)
plt.xlim([0,5/ff]) #Para que me muestre 5 ciclos
plt.plot(tt, xx, marker = 'X', color = 'r', linestyle = '-')
plt.grid()
plt.title("Salida Generador de Senales")
plt.ylabel("Amplitud [V]")
plt.xlabel("Tiempo [s]")
```
